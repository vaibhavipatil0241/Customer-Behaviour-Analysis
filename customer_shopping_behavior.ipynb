```python
import pandas as pd

df = pd.read_csv('customer_shopping_behavior.csv')
```

```python
# Imputing missing values in Review Rating column with the median rating of the product category

df['Review Rating'] = df.groupby('Category')['Review Rating'].transform(lambda x: x.fillna(x.median()))
```

```python
df.isnull().sum()
```

```python
# Renaming columns according to snake casing for better readability and documentation

df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ','_')
df = df.rename(columns={'purchase_amount_(usd)':'purchase_amount'})
```

```python
df.columns
```

```python
# for creating a new column age_group
labels = ['Young Adult', 'Adult', 'Middle-aged', 'Senior']
df['age_group'] = pd.qcut(df['age'], q=4, labels = labels)
```

```python
df[['age','age_group']].head(10)
```

```python
# create new column purchase_frequency_days

frequency_mapping = {
    'Fortnightly': 14,
    'Weekly': 7,
    'Monthly': 30,
    'Quarterly': 90,
    'Bi-Weekly': 14,
    'Annually': 365,
    'Every 3 Months': 90
}

df['purchase_frequency_days'] = df['frequency_of_purchases'].map(frequency_mapping)
```

```python
df[['purchase_frequency_days','frequency_of_purchases']].head(10)
```

```python
df[['discount_applied','promo_code_used']].head(10)
```

```python
(df['discount_applied'] == df['promo_code_used']).all()
```

```python
# Dropping promo code used column

df = df.drop('promo_code_used', axis=1)
```

```python
from sqlalchemy import create_engine
import psycopg2
from psycopg2 import sql

# Step 1: Connection details
username = "postgres"
password = "Aadhusql"
host = "localhost"
port = "5432"
database = "customer_behavior"

# Step 2: Connect to default database (postgres) first
default_db = "postgres"
conn = psycopg2.connect(
    dbname=default_db, user=username, password=password, host=host, port=port
)
conn.autocommit = True
cur = conn.cursor()

# Step 3: Check if database exists, if not, create it
cur.execute("SELECT 1 FROM pg_database WHERE datname = %s;", (database,))
exists = cur.fetchone()
if not exists:
    cur.execute(sql.SQL("CREATE DATABASE {}").format(sql.Identifier(database)))
    print(f"Database '{database}' created successfully.")
else:
    print(f"Database '{database}' already exists.")

cur.close()
conn.close()

# Step 4: Connect to the newly created or existing database
engine = create_engine(f"postgresql+psycopg2://{username}:{password}@{host}:{port}/{database}")

# Step 5: Load DataFrame into PostgreSQL
table_name = "customer"
df.to_sql(table_name, engine, if_exists="replace", index=False)

print(f"✅ Data successfully loaded into table '{table_name}' in database '{database}'.")
```
